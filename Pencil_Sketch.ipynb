# New Section

```python


def pecile_sketch(img_path, output_path):
    img = cv2.imread(img_path)
    print("Original Image")
    cv2_imshow(img)
    grey_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    inverted_grey_img = cv2.bitwise_not(grey_img)
    blurred_img = cv2.GaussianBlur(inverted_grey_img, (21, 21), 0)

    inverted_blurred_img = cv2.bitwise_not(blurred_img)
    pencil_sketch_img = cv2.divide(grey_img, inverted_blurred_img, scale=256.0)

    cv2.imwrite(output_path, pencil_sketch_img)
    # Use cv2_imshow instead of cv2.imshow
    print("Pencil Sketch Image")
    cv2_imshow(pencil_sketch_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

input_img_path = "/content/images.jpg"
output_img_path = "naruto.jpg"
pecile_sketch(input_img_path, output_img_path)
```
